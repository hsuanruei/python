```python
import matplotlib.pyplot as plt

def main():
    x_coords = [0 ,1 ,2 ,3 ,4 ]
    y_coords = [0 ,3 ,1 ,5 ,2 ]
   
    plt.plot(x_coords,y_coords, marker= 'o')
    
    plt.xlim(xmin=-5,xmax=10)
    plt.ylim(ymin=-5,ymax=10)
    plt.title('Sales by Year')
    plt.xticks(x_coords,['2016','2017','2018','2019','2020'])
    plt.yticks(x_coords,['$1m','$2m','$3m','$4m','$5m'])
    plt.xlabel('Year')
    plt.ylabel('Sales')
    plt.grid(True)
    
    plt.show()
    
main()
```

```python
將X軸變成年分
y軸便銷售額
手動圖表大小 xlim & ylim
```

```python
import matplotlib.pyplot as plt

x_coords = [10,20,30,40,50]
heights = [100,200,300,400,500]

bar_width =5

plt.bar(x_coords,heights, bar_width, color=('r','g','b','c','k'))

plt.title('Sales by Year')
plt.xlabel('Year')
plt.ylabel('Sales')

plt.xticks(x_coords, ['2016','2017','2018','2019','2020'])
plt.yticks(heights,['$1m','$2m','$3m','$4m','$5m'])
plt.show()
```

```python
import matplotlib.pyplot as plt

sales = [100,400,300,600]
slice_labels = ['1st Qtr', '2nd Qtr','3rd Qtr','4th Qtr']
separated = (.2,0,.5,0)


plt.pie(sales, labels=slice_labels, autopct='%.2f%%',explode=separated, shadow=True)
plt.title('Sales by Quarter')

plt.show()
```

圓餅圖
 shadow=True增加陰影，更有立體感
 explode=separated拆分
 autopct='%.2f%%'顯現百分比
